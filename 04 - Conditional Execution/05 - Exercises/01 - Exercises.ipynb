#### Exercise 3

Given an credit score `score`, assign a string value to another variable `rating` based on the following scale:

- [0, 580) --> Poor
- [580, 670) --> Fair
- [670, 740) --> Good
- [740, 800) --> Very Good
- [800, 850] --> Excellent


```python
score = 581

if score <= 580:
    result = "Poor"
elif 580 < score <= 670:
    result = "Fair"
elif 670 < score <= 740:
    result = "Good"
elif 740 < score <= 800:
    result = "Very Good"
elif 800 < score <= 850:
    result = "Excellent"
else:
    result = "Invalid score"

print(result)
```
